```python
import random

# Game title
gameTitle = "WORD GUESSING"

# Work bank list to store words from file
wordBankList = []

# Open file in read mode
with open("./sample_data/words.txt") as file:
  for line in file:
    wordBankList.append(line.rstrip().lower())

# Choose a random word from the word bank list
wordToGuess = (random.choice(wordBankList)).lower()

# Game variables definition
maxTurns = 5
misplacedLetters = []
incorrectLetters= []
turnsTakenByPlayer = 0
wordGuessLength = 5

# Print welcome message and game expectations
print("*"*100)
print(f"                              WELCOME TO THE {gameTitle} GAME                               ")
print("*"*100)
print("In this game, you will have to guess the 5-letter word that is randomly generated by the computer.")
print(f"You have {maxTurns} attempts to guess this word.")
print("\n")

# Start game loop
while turnsTakenByPlayer < maxTurns:
    playerWordGuess = input("Enter your guess here --> ")
    playerWordGuess = playerWordGuess.lower()

    correctLetters = 0
    print("")

    # Make sure length of word guessed is 5 and make sure word guessed contain letters
    if len(playerWordGuess) != len(wordToGuess) or not playerWordGuess.isalpha():
        print("Word is more or less than 5. Please re-enter a 5-letter word.")
        continue

    # Check and put incorrect letters of the guess in a list
    # Check and put misplaced letter of the guess in a list
    for letterPosition in range(wordGuessLength):
        if playerWordGuess[letterPosition] == wordToGuess[letterPosition]:
          if playerWordGuess[letterPosition] in misplacedLetters:
            misplacedLetters.remove(playerWordGuess[letterPosition])
          print(playerWordGuess[letterPosition], end="")
          correctLetters+=1
        elif playerWordGuess[letterPosition] != wordToGuess[letterPosition]:
            if playerWordGuess[letterPosition] not in misplacedLetters:
                misplacedLetters.append(playerWordGuess[letterPosition])

            if playerWordGuess[letterPosition] not in wordToGuess:
              if playerWordGuess[letterPosition] not in incorrectLetters:
                  incorrectLetters.append(playerWordGuess[letterPosition])
              print("_ ", end="")
            else:
              correctLetters+=1
              print("_ ", end="")

    turnsTakenByPlayer+=1

    # Notifications
    if playerWordGuess == wordToGuess:
        print("\n\nCongratulations, you won!!! You guessed the word right.")
        break
    elif turnsTakenByPlayer == maxTurns:
        print("\n\nSorry, you lost. Try again.")
        print(f"The word was {wordToGuess}.")
        break
    else:
        print(f"\n\nYou guessed {correctLetters} letter(s) correctly.")

        if len(incorrectLetters) > 0:
          print("\nLetters you guessed incorrectly are ", end="")
          print( *incorrectLetters, sep=', ')

        if len(misplacedLetters) > 0:
          print("\nLetters you misplaced are ", end="")
          print(*misplacedLetters, sep=', ')

        print(f"\nYou have {maxTurns - turnsTakenByPlayer} attempt(s) left to play the game.\n")
```
